```python
# Specify Pattern Using RegEx: To specify regular expressions, metacharacters are used. 
#--------------------------------------------------------------------------------------------------------------
# MetaCharacters: 

#characters that are interpreted in a special way by a RegEx engine. Here's a list of metacharacters:
#[] . ^ $ * + ? {} () \ |

#[] - Square brackets: Square brackets specifies a set of characters you wish to match.
#Example: [abd-z] will match if the string you are trying to match contains any of the a, b or c

#You can also specify a range of characters using - inside square brackets.
#[a-e] is the same as [abcde].
#[1-4] is the same as [1234].
#[0-39] is the same as [01239].

'[a,A].+labs'   - Any word satisfies the criteria start with a or A and ends with labs and between at least one character should be there
'[a,A].*labs'   - Any word satisfies the criteria start with a or A and ends with labs

re.findall('[a,A].+labs'. 'alabs') = won't satisfy the criterial
re.findall('[a,A].*labs'. 'alabs') = satisfy the criterial

re.findall('[a,A].+labs'. 'anlabs') = satisfy the criteria
re.findall('[a,A].*labs'. 'anlabs') = satisfy the criteria


#You can complement (invert) the character set by using caret ^ symbol at the start of a square-bracket.
#[^abc] =  means any character except a or b or c.
#[^0-9] means any non-digit character.

#Period(.): A period matches any single character (except newline '\n').

#Caret(^): The caret symbol ^ is used to check if a string starts with a certain character.
#Dollar($): The dollar symbol $ is used to check if a string ends with a certain character.

#star(*): The star symbol * matches zero or more occurrences of the pattern left to it.
#Plus(+): The plus symbol + matches one or more occurrences of the pattern left to it.
#Question Mark(?): The question mark symbol ? matches zero or one occurrence of the pattern left to it.

#Braces({}): Consider this code: {n,m}. This means at least n, and at most m repetitions of the pattern left to it.
            #Let's try one more example. This RegEx [0-9]{2, 4} matches at least 2 digits but not more than 4 digits
#Alternation(|): Vertical bar | is used for alternation (or operator).
            #a|b match any string that contains either a or b
#Group(()):Parentheses () is used to group sub-patterns. For example, (a|b|c)xz match any string that matches either a or b or c followed by xz
#Backslash(\): Backlash \ is used to escape various characters including all metacharacters. 
            #For example,\$a match if a string contains $ followed by a. Here, $ is not interpreted by a RegEx engine in a special way.
            #If you are unsure if a character has special meaning or not, you can put \ in front of it. This makes sure the character is not treated in a special way.
#---------------------------------------------------------------------------------------------------------------------------------

#Special Sequences: Special sequences make commonly used patterns easier to write. 
#Here's a list of special sequences:

#\A - Matches if the specified characters are at the start of a string.
#\b - Matches if the specified characters are at the beginning or end of a word.
#\B - Opposite of \b. Matches if the specified characters are not at the beginning or end of a word.
#\d - Matches any decimal digit. Equivalent to [0-9]
#\D - Matches any non-decimal digit. Equivalent to [^0-9]
#\s - Matches where a string contains any whitespace character. Equivalent to [ \t\n\r\f\v].
#\S - Matches where a string contains any non-whitespace character. Equivalent to [^ \t\n\r\f\v].
#\w - Matches any alphanumeric character (digits and alphabets). Equivalent to [a-zA-Z0-9_]. By the way, underscore _ is also considered an alphanumeric character.
#\W - Matches any non-alphanumeric character. Equivalent to [^a-zA-Z0-9_]
#\Z - Matches if the specified characters are at the end of a string.

#---------------------------------------------------------------------------------
#Special methods in regular expression module
#re.compile()
#re.findall() method returns a list of strings containing all matches.
#re.split method splits the string where there is a match and returns a list of strings where the splits have occurred.
#re.sub() is The method returns a string where matched occurrences are replaced with the content of replace variable.
#re.subn() is similar to re.sub() expect it returns a tuple of 2 items containing the new string and the number of substitutions made.
#re.search() method looks for the first location where the RegEx pattern produces a match with the string.
#match.group() method returns the part of the string where there is a match.
#match.start(), match.end() and match.span(): The start() function returns the index of the start of the matched substring. Similarly, end() returns the end index of the matched substring.
#match.re and match.string: re attribute of a matched object returns a regular expression object. Similarly, string attribute returns the passed string.
```

```python
# The match function in re
# Matches for a pattern at the begining of a string

line1 = "this is PM modi" # no match
line2 = "Modi wins the election" # matches here
# if the pattern is 'modi'

# This is a literal form of regex
# r'string' --> raw text : text that can't be processed
matchObj = re.match( r'modi', line1, re.I)
# re.I : Case Insensitive matching

if matchObj:
    print ("Match found at the begining of the string!!")
    print ("matchObj.group() : ", matchObj.group())
else:
    print ("No match!!")
```

```python
matchObj = re.match( r'modi', line2, re.M|re.I)

if matchObj:
    print ("Match found at the begining of the string!!")
    print ("matchObj.group() : ", matchObj.group())
else:
    print ("No match!!")
```

```python
# The search function in re
# Matches for a pattern anywhere in a string

line1 = "this is PM modi"
line2 = "Modi wins the election"

matchObj = re.search( r'modi', line1, re.I)

if matchObj:
    print ("Match found in the string!!")
    print ("matchObj.group() has found : ", matchObj.group())
else:
    print ("No match!!")
matchObj
matchObj.group()
```

### Quantifiers

These symbols convey "how many times" a pattern has to be present inside a string. Whether it is present atleast once or multiple or exactly once or exactly n times etc.

#### 1. Exactly one occurance or none at all `?`

There will be a match if a pattern is present ___exactly once___ or __it is not present at all__

```python
if re.search(pattern="\d?", string="abcdef"):
    print("match for pattern is found!")
else:
    print("no match")
```

```python
if re.search(pattern="\d?", string="abc8def"):
    print("match for pattern is found!")
else:
    print("no match")
```

```python
s1 = "9gag"
s2 = "marathon"

pat = r'[0-9]?'
# search for exactly one number or none at all
# Both s1 and s2 should match

searchObj = re.search( r'[0-9]?', s1) 

if searchObj:
    print ("Matched Objects for s1: ", searchObj.group())
else:
    print ("No match at all!!")

#---------------------------------------------------------
searchObj = re.search( r'[0-9]?', s2) 

if searchObj:
    print("-----------------------------------------------")
    print ("Matched Objects for s2: ", searchObj.group())
    print("Not going into unmatched state")
else:
    print ("No match at all!!")
```

#### 2. Atleast one occurance `+`

There will be a match if a pattern is present ___at least once___

#### 1. `^` : start of the string

This is differnet from caret when used inside a bracket `[^]`.

```python
# this is different from [^] 

# Aim : to search for the pattern "python" in the start of the string

s1 = "we are learning python for data science"
s2 = "python is fun"

searchObj = re.search( r'^python', s1) 
if searchObj:
    print ("Matched Objects for s1: ", searchObj.group())
else:
    print ("No match at all!!")

#---------------------------------------------------------
searchObj = re.search( r'^python', s2) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s2: ", searchObj.group())
else:
    print ("No match at all for s2!!")
```

#### 2. `$` : end of the string

```python

# Aim : to search for the pattern "python" in the start of the string

s1 = "we are learning python for data science"
s2 = "python is fun"
s3 = "I love Python"

searchObj = re.search( r'python$', s1, re.I) 
if searchObj:
    print ("Matched Objects for s1: ", searchObj.group())
else:
    print ("No match at all!!")

#---------------------------------------------------------
searchObj = re.search( r'python$', s2, re.I) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s2: ", searchObj.group())
else:
    print ("No match at all for s2!!")

#---------------------------------------------------------
searchObj = re.search( r'python$', s3, re.I) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s3: ", searchObj.group())
else:
    print ("No match at all for s3!!")
```

#### 3. `\bxxx\b` : Exact pattern within a string.

This would match the exact pattern but won't match for strings that contain the given patten along with something else

```python
s1 = "This is a good vapo rub"
s2 = "Malaysia has largest rubber production"
s3 = "Brabbbruby is an easy language"

searchObj = re.findall( r'rub', s1) 
if searchObj:
    print ("Matched Objects for s1: ", searchObj)
else:
    print ("No match at all!!")

#---------------------------------------------------------
searchObj = re.findall(r'rub', s2) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s2: ", searchObj)
else:
    print ("No match!!")

#---------------------------------------------------------
searchObj = re.findall(r'rub', s3) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s3: ", searchObj)
else:
    print ("No match!!")
```

#### 4. `\bxxx\B` : Sting containing the pattern but not the exact pattern

This is the exact opposite of the above boundary patterm. This would match for the strings that contain the given patten along with something else but will __never__ match if the patten is present alone

```python
s1 = "rub"
s2 = "rubber"
s3 = "ruby"

searchObj = re.findall( r'\brub\B', s1) 
if searchObj:
    print ("Matched Objects for s1: ", searchObj)
else:
    print ("No match at all!!")

#---------------------------------------------------------
searchObj = re.match(r'\brub\B', s2) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s2: ", searchObj.group())
else:
    print ("No match!!")

#---------------------------------------------------------
searchObj = re.match(r'\brub\B', s3) 
print("-----------------------------------------------")
if searchObj:
    print ("Matched Objects for s3: ", searchObj.group())
else:
    print ("No match!!")
```

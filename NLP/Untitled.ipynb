```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('reviews.csv')
```

```python
print("Dataset Info:")
print(df.info())
```

```python
print("\nFirst 5 Rows:")
print(df.head())
```

```python
print("\nSummary Statistics:")
print(df.describe(include='all'))
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='stars', data=df, palette='viridis')
plt.title('Distribution of Star Ratings')
plt.xlabel('Stars')
plt.ylabel('Number of Reviews')
plt.show()
```

```python
most_reviewed_business = df['business_id'].value_counts().idxmax()
most_reviews = df['business_id'].value_counts().max()
print(f"\nMost reviewed business: {most_reviewed_business} with {most_reviews} reviews")
```

```python
most_active_user = df['user_id'].value_counts().idxmax()
most_reviews_user = df['user_id'].value_counts().max()
print(f"Most active user: {most_active_user} with {most_reviews_user} reviews")
```
